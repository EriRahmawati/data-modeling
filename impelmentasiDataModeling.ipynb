```python
c = conn.cursor()
```

```python
c.executescript('drop table if exists country')
c.executescript('drop table if exists product')
c.executescript('drop table if exists costumer')
c.executescript('drop table if exists pemesanan')
c.execute("CREATE TABLE IF NOT EXISTS 'country' ('id' INT, 'country_name' VARCHAR(255) NULL, 'state' VARCHAR(255) NULL, 'City' VARCHAR(255), 'region' VARCHAR(255) NULL, 'postal_code' INT NULL, PRIMARY KEY ('id'))")
c.execute("CREATE TABLE IF NOT EXISTS 'product' ( 'id' INT, 'product_name' VARCHAR(255) NULL, 'sales' FLOAT NULL, 'category' VARCHAR(255) NULL, 'sub_category' VARCHAR(255) NULL, 'discount' FLOAT NULL, 'profit' FLOAT NULL,  PRIMARY KEY ('id'))")
c.execute("CREATE TABLE IF NOT EXISTS 'costumer' ('id' INT, 'costumer_name' VARCHAR(255) NULL, 'segment' VARCHAR(255) NULL, 'country_id' INT NOT NULL, PRIMARY KEY ('id'), FOREIGN KEY ('country_id') REFERENCES 'country' ('id') ON DELETE NO ACTION ON UPDATE NO ACTION)")
c.execute("CREATE TABLE IF NOT EXISTS 'pemesanan' ( 'id' INT, 'order_date' DATE NULL, 'Quantity' INT NULL, 'ship_mode' VARCHAR(255) NULL, 'ship_date' DATE NULL, 'product_id' INT NOT NULL, 'costumer_id' INT NOT NULL, PRIMARY KEY ('id'), FOREIGN KEY ('product_id') REFERENCES 'product' ('id') ON DELETE NO ACTION ON UPDATE NO ACTION, FOREIGN KEY ('costumer_id') REFERENCES 'costumer' ('id') ON DELETE NO ACTION ON UPDATE NO ACTION)")
```

```python
df = pd.read_excel("US Superstore data.xls")
df
```

```python
country = []
product = []
costumer = []
pemesanan=[]


for i,j in df.iterrows():
    if j[8] not in country:
        country.append(j[8])
        
    k = -1
    for index,values in enumerate(country):
        if values == j[8]:
            k = index+1
            break
            
for i,j in df.iterrows():
    if j[13] not in product:
        product.append(j[13])
        
    p = -1
    for index,values in enumerate(product):
        if values == j[13]:
            p = index+1
            break
            

    
for i,j in df.iterrows():
    if j[5] not in costumer:
        costumer.append(j[5])
        
    c = -1
    for index,values in enumerate(costumer):
        if values == j[5]:
            c = index+1
            break
                  
if[j[1], j[2], j[18], j[4], j[3], p, c] not in pemesanan:
    pemesanan.append([j[1], j[2], j[18], j[4], j[3], p, c])
    
    
print(country)
print(product)
print(costumer)
print(pemesanan)
```
